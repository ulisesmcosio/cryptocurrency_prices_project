```python
"""
In this project we analyzed the price of various cryptocurrencies such as Bitcoin, Ethereum, etc. and
compared them to one another using various visualization techniques in python using numpy, pandas,
matplotlib and seaborn.
"""

# First we are going to import the necessary packages into our jupyter notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Next, we are going to import the CSV files with our crypto daily prices and daily returns using pandas

prices_df = pd.read_csv('crypto_daily_prices.csv')
returns_df = pd.read_csv('crypto_daily_returns.csv')

prices_df
```

```python
# Daily prices for different cryptocurrencies in a single plot

prices_df.plot(x = 'Date', y = ['BTC-USD Price','ETH-USD Price','LTC-USD Price'], figsize = (15, 10))
plt.title('Cryptocurrency Daily Prices')
plt.ylabel('Price [$]')
plt.grid()
```

```python
# Cryptocurrency daily prices shown in multiple subplots

plt.figure(figsize = (20, 10))

# Here we plot the first subplot
plt.subplot(2, 2, 1)
plt.plot(prices_df['BTC-USD Price'], 'b')
plt.title('Bitcoin')
plt.grid()

# Code for 2nd plot
plt.subplot(2, 2, 2)
plt.plot(prices_df['ETH-USD Price'], 'r')
plt.title('Ethereum')
plt.grid()

# Code for 3rd plot
plt.subplot(2, 2, 3)
plt.plot(prices_df['LTC-USD Price'], 'g')
plt.title('Litecoin')
plt.grid()
```

```python
# Next, we wanted to compare visually the daily returns between Bitcoin and Ethereum
# To do this first, we reminded ourselves of what the daily returns df looks like

returns_df

# Scatterplot between Bitcoin and Ethereum daily returns

plt.figure(figsize = (15, 10))
plt.scatter(returns_df['BTC'], returns_df['ETH'])
plt.grid()
plt.title('Scatterplot of BTC vs. Ethereum')
```

```python
'''
I personally believe Seaborn has stunning visualizations, so here is the same scatterplot but using the Seaborn
package
'''
plt.figure(figsize = (10, 10))
sns.scatterplot(data = returns_df, x = 'BTC', y = 'ETH')
```

```python
'''Pie chart showing crypto portfolio with 60% XRP, and the rest equally divided.
Investment in Bitcoin was exploded for emphasis.
'''

values = [13.33, 13.33, 13.33, 60]
colors = ['y', 'r', 'b', 'm']
explode = [0, 0, 0, 0.2]
labels = ['BTC', 'ETH', 'LTC','XRP']

plt.figure(figsize = (8, 8))
plt.title('Crypto Portfolio')
plt.pie(values, colors = colors, labels = labels, explode = explode)
plt.grid()
```

```python
# Histogram for ETH daily returns, with 30 bins and red in color

# First, we need to calculate the mean of our data frame
mu = returns_df['ETH'].mean()
# Next, we need to calculate the standard deviation
sigma = returns_df['ETH'].mean()

num_of_bins = 30

plt.figure(figsize = (15, 10))
plt.hist(returns_df['ETH'], num_of_bins, facecolor = 'blue')
plt.title('Histogram for ETH Daily Returns')

# Although we won't label the histogram with mu & sigma, we can always come back to this cell and do it in a matter
# seconds. This was done to avoid clutter and provide a cleaner visualization.
```

```python
# Finally, we want to show the correlations between the daily returns of the cryptocurrencies. We're going
# to do this using a heat map in Seaborn

plt.figure(figsize = (10, 10))
sns.heatmap(returns_df.corr(), annot = True)
```
